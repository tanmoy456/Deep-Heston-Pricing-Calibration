```python
import gzip
import numpy as np
import matplotlib.pyplot as plt

def fun_(mode='P1P2'):

    if mode=='NN-Stochstic-Vol':
        file = gzip.GzipFile('HestonTrainSet.txt.gz', 'r')
    elif mode=='P1P2':
        file = 'Heston_P1P2_LHS_Feller_CALL_r_0.0_T_K_20k.npy'
    elif mode=='ql':
        file = 'Heston_ql_LHS_Feller_CALL_r_0.0_T_K_20k.npy'
    elif mode=='COS':
        file = 'Heston_COS_LHS_Feller_CALL_r_0.0_T_K_20k.npy'
        
    data = np.load(file)
    params_ref = data[:, :5]
    
    param_names = ['$v_0$', '$\\rho$', '$\\sigma$', '$\\theta$', '$\\kappa$']
    
    plt.figure(figsize=(12, 6))
    for i, name in enumerate(param_names):
        plt.subplot(2, 3, i+1)
        plt.hist(params_ref[:, i], bins=50, alpha=0.7, edgecolor='black')
        plt.title(f"Histogram of {name}")
        plt.xlabel(name)
        plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
fun_(mode='NN-Stochstic-Vol')
```
